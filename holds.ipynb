## II. Count some stuff

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000)

array = moon.data.problems_to_array(probs)
counts_bytype = np.sum(array, axis=0)
counts = np.sum(counts_bytype[:,:,[0,1,2]], axis=-1)
```

```python
usage_bytype = counts_bytype.astype(np.float32)
for i in range(0, 4):
    usage_bytype[:,:,i] /= np.sum(usage_bytype[:,:,i])

display(usage_bytype[:,:,0])  # End holds (red)
display(usage_bytype[:,:,1])  # Start holds (green)
display(usage_bytype[:,:,2])  # Intermediate holds (blue)
display(usage_bytype[:,:,3])  # Not-used holds
```
