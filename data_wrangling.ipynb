### Selection, formatting

```python
cols2keep = ['Name', 'Grade', 'UserGrade', 'Setter', 'UserRating', 'Repeats', 'IsBenchmark',
             'IsMaster', 'IsAssessmentProblem', 'ProblemType', 'Moves', 'DateTimeString']

probs_clean = problems[cols2keep].copy()  # We do copy() because of column-adding operations next
probs_clean.sample(5)
```

Only thing is that we want to expand on those deeper cells still (Moves and Setter).

```python
# Expand on Setter info
setters = probs_clean['Setter']
keys = ['Firstname', 'Lastname', 'Nickname', 'City', 'Country']
for k in keys:
    vals = [s[k] for s in setters]
    probs_clean['Setter.%s' % k] = vals

# Expand on Moves info: filter moves by start, intermediate, end
moves = probs_clean['Moves']
def _moves2holdcols(dicts):
    start, intermed, end = [], [], []
    for d in dicts:
        if d['IsStart']: start.append(d['Description'].upper())
        elif d['IsEnd']: end.append(d['Description'].upper())
        else: intermed.append(d['Description'].upper())
    return [start, intermed, end]

# Add columns for holds
hold_cols = [_moves2holdcols(move_dicts) for move_dicts in moves]
probs_clean['Holds.Start'] = [h[0] for h in hold_cols]
probs_clean['Holds.Intermed'] = [h[1] for h in hold_cols]
probs_clean['Holds.End'] = [h[2] for h in hold_cols]

# Remove unexpanded cols
probs_clean.drop('Setter', axis=1, inplace=True)
probs_clean.drop('Moves', axis=1, inplace=True)

display(probs_clean.tail())
print(len(probs_clean))
```

### Deduplication

Getting close! Things are falling into place, but we should check for dups.

Interestingly, there are some some problems that are identical in terms of holds but differ in terms of name, and often grade. We don't prune those; just those that are obviously accidental dups.

```python
dups = probs_clean[probs_clean.applymap(lambda x: str(sorted(x)) if isinstance(x, list) else x).duplicated(['Holds.Start', 'Holds.Intermed', 'Holds.End'], keep=False)]
display(dups.sort_values('Holds.Start'))

names = dups[['Name', 'Setter.Nickname']]
dup_names = names[names.duplicated()]
display(dup_names)  # I'm not sure why some problems are still represented multiple times here; they shouldn't be
```

```python
keeps, drops = set(), set()

for _, (name, setter) in dup_names.iterrows():
    # Get all probs that have the same name and setter
    same_probs = dups[(dups['Name'] == name) & (dups['Setter.Nickname'] == setter)]
    repeats, ratings = same_probs['Repeats'], same_probs['UserRating']
    # Prefer higher repeats, then higher ratings, and otherwise newest
    if (repeats.iloc[0] != repeats.iloc[1:]).any():
        keep = repeats.idxmax()
    elif (ratings.iloc[0] != ratings.iloc[1:]).any():
        keep = ratings.idxmax()
    else:
        keep = same_probs.index[-1]
    # Keep the preferred prob and drop the rest
    keeps.add(keep)
    drops.update(set(same_probs.index) - set([keep]))
```

```python
print('Keeping {} problems and dropping {}'.format(len(keeps), len(drops)))

probs_dedup = probs_clean.drop(drops).reset_index(drop=True)
len(probs_dedup)
```

## Downgrade Problems

Technically, the lowest grade in the MB2016 dataset is 6B+. However, some problems have names which suggest lower grades (6A+, 6A, ...). Let's change those.

Interestingly, other problems which are also 6B+ have *higher* grades in their names. For these, however, it's unclear whether the setter was just being silly (and giving it an improper name) or lazy (and actually not changing the grade from 6B+). We don't do anything with these, even though it might be irksome, and there are a lot of 6B+ problems in this dataset.

The other issue here is that some problems with the grade in their name could actually be referring to V-grade instead of Font grade. Unfortunately, we just don't have the information to make sure all problems are labeled correctly, We'll assume Font grade unless V-grade is implied.

```python
fives = ['5A', '5B', '5C', '5D', '5\+']
sixes = ['6A', '6A\+', '6B']

total = 0
for font in fives:
    candidates = probs_dedup['Name'].str.contains(font)
    vs = probs_dedup[candidates]['Name'].str.contains('V5')
    likely_fonts = probs_dedup[candidates & ~vs]
    display(likely_fonts['Name'])
    total += len(likely_fonts)
    
for font in sixes:
    candidates = probs_dedup['Name'].str.contains(font)
    vs = probs_dedup[candidates]['Name'].str.contains('V6')
    if '+' in font:
        likely_fonts = probs_dedup[candidates & ~vs]
    else:
        upgrades = probs_dedup[candidates]['Name'].str.contains(font + '\+')
        likely_fonts = probs_dedup[candidates & ~vs & ~upgrades]
    display(likely_fonts['Name'])
    total += len(likely_fonts)
    
print(total)
```

It looks like these are all the problems we want to target for a grade change---158 of them! This is not a huge number compared to the whole dataset; however, it is somewhat of a huge addition to grossly unbalanced lower grades.

I had worried that some problems might have grade-like strings in their names (but not actually be indicating a grade)... but it doesn't seem like that's an issue for us.

There are some problems that list multiple grades: like "5C 6A" or what have you. Eh, we'll arbitrarily take the highest grade that the above search method detects.

```python
import numpy as np

probs_regrade = probs_dedup.copy()

for font in fives:
    candidates = probs_dedup['Name'].str.contains(font)
    vs = probs_dedup[candidates]['Name'].str.contains('V5')
    idxs = candidates & ~vs
    # Regrade
    probs_regrade.loc[idxs, 'Grade'] = '5+'
    probs_regrade.loc[idxs, 'UserGrade'] = np.nan
    
for font in sixes:
    candidates = probs_dedup['Name'].str.contains(font)
    vs = probs_dedup[candidates]['Name'].str.contains('V6')
    idxs = candidates & ~vs
    if '+' not in font:
        upgrades = probs_dedup[candidates]['Name'].str.contains(font + '\+')
        idxs = idxs & ~upgrades
    # Regrade
    probs_regrade.loc[idxs, 'Grade'] = font.replace("\\", '')
    probs_regrade.loc[idxs, 'UserGrade'] = np.nan
```
