## I. Checking Out Hold Usage
### I.a A Few (random) Example Problems

```python
import moon.viz

array = moon.data.problems_to_array(probs)
moon.viz.array_to_image(array[100], show=True)
moon.viz.array_to_image(array[400], show=True)
moon.viz.array_to_image(array[800], show=True)
```

### I.b Countin' Some Stuff

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000)

counts_bytype = np.sum(array, axis=0)
counts = np.sum(counts_bytype[:, :, [0,1,2]], axis=-1)

usage_bytype = counts_bytype.astype(np.float32)
for i in range(0, 4):
    usage_bytype[:,:,i] /= np.sum(usage_bytype[:,:,i])

display(usage_bytype[:,:,0])  # End holds (red)
display(usage_bytype[:,:,1])  # Start holds (green)
display(usage_bytype[:,:,2])  # Intermediate holds (blue)
display(usage_bytype[:,:,3])  # Not-used holds
```

```python
usage = counts.astype(np.float32) / np.sum(counts)
display(usage)
```

### I.b Visualizin' Hold Usage

```python
usage_img = moon.viz.usage_to_image(usage, 'P')
e_usage_img = moon.viz.usage_to_image(usage_bytype[:,:,0], 'R')
s_usage_img = moon.viz.usage_to_image(usage_bytype[:,:,1], 'G')
o_usage_img = moon.viz.usage_to_image(usage_bytype[:,:,2], 'B')
```
